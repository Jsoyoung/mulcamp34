```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)

ax[0].set_title("ax.[0] title") # 첫 번째 차트에 제목 생성
ax[0].plot([1,2,3]) # 첫 번째 차트 생성

ax[1].set_title("ax.[1] title")  # 두 번째 차트에 제목 생성
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax = ax[1], data = tips)  # 두 번째 차트 생성

# plt.title("plt.title") # 한 쪽 차트에만 들어감. 쓸 수 X
fig.suptitle("fig title")  # 가운데에 제목 생성

plt.show()

# 중고급 레벨..: https://matplotlib.org/stable/users/explain/axes/arranging_axes.html
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)  # 2*2 배열 차트일 때는 (0,0)(0,1)(1,0)(1,1)형식!


ax[0, 0].set_title("ax.[0] title")
ax[0, 0].plot([1,2,3])

ax[1, 1].set_title("ax.[1] title")
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'sex', ax = ax[1, 1], data = tips)


fig.suptitle("fig title")
plt.tight_layout()  # 타이트한 차트의 간격이 벌어짐
plt.show()
```

## 산점도 그래프

## 시각화 중급 레벨

### 라이브러리 가져오기

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator, FuncFormatter)  # 축 설정할 때 사용하는 값
import seaborn as sns
import numpy as np
```

### 데이터 불러오기

```python
tips = sns.load_dataset("tips")
tips.head(3)
```

```python
# 사용자 정의함수 추가
# y축 설정할 때, 값을 소수점 2자리까지 표현
def major_formatter(x, pos):
    return "%.2f$" % x
formatter = FuncFormatter(major_formatter)
```

### 객체 지향 선언

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (16, 5))


# 이상적인 막대 그래프
ax0 = sns.barplot(x = "day", y = 'total_bill', data = tips, errorbar=None, color='lightgray',
                  zorder=2, alpha=0.85,ax=ax[0]) # zorder=2 그래프 선을 바 뒤로 보내기

group_mean = tips.groupby(['day'])['total_bill'].agg('mean')
# print(group_mean)
h_day = group_mean.sort_values(ascending=False).index[0]  # 해당되는 결과값의 최대값(MAX)의 인덱스 요일을 하나 가져옴
h_mean = np.round(group_mean.sort_values(ascending=False)[0], 2)  # 해당되는 결과값의 MAX값을 하나 가져옴
for p in ax0.patches:
  fontweight = "normal"
  color = "k"
  height = np.round(p.get_height(), 2)
  if h_mean == height:
    fontweight="bold"
    color="darkred"  # Adobe Color Wheel 색상코드 추가
    p.set_facecolor(color)  # 옵션 추가
    p.set_edgecolor("black")  # 옵션 추가
  ax0.text(p.get_x() + p.get_width()/2., height+1, height, ha = 'center',
           size=12,
           fontweight=fontweight, color=color)  #옵션 추가(font, size, color 등등)

# matplotlib.patches.Rectangle에서 참고하여 옵션 적용할 수 있음

# y축 범위 지정 옵션
ax0.set_ylim(-3, 30)
ax0.set_title("Ideal Bar Graph", size = 16)

# 가로세로 축, 일부 수정 및 변경
ax0.spines['top'].set_visible(False)
ax0.spines['left'].set_position(("outward", 20))
ax0.spines['left'].set_visible(False)
ax0.spines['right'].set_visible(False)

# y축 값 변경
ax0.yaxis.set_major_locator(MultipleLocator(10))
ax0.yaxis.set_major_formatter(formatter)
ax0.yaxis.set_minor_locator(MultipleLocator(5))

# y축 라벨 변경
ax0.set_ylabel("Avg. Total Bill($)", fontsize=14)

# y축 Grid 라인 수정
ax0.grid(axis="y", which="major", color="lightgray")
ax0.grid(axis="y", which="minor", ls=":")

# x축의 값을 변경
ax0.set_xlabel("Weekday", fontsize=14)
for xtick in ax0.get_xticklabels():
  print(xtick)
  if xtick.get_text() == h_day:
    xtick.set_color("darkred")
    xtick.set_fontweight("demibold")
ax0.set_xticklabels(['Thursday', 'Friday', 'Saturday', 'Sunday'], size=12)



# 그냥 일반 막대 그래프
ax1 = sns.barplot(x = "day", y = 'total_bill', data = tips, errorbar=None, alpha=0.85, ax=ax[1])
# errorbar=None : 에러 바 안보이게 하기
for p in ax1.patches:
  height = np.round(p.get_height(), 2)
  ax1.text(p.get_x() + p.get_width()/2., height+1, height, ha = 'center', size=12)
ax1.set_ylim(-3, 30)
ax1.set_title("Just Bar Graph")


plt.show()
```
