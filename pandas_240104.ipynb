```python
iris = sns.load_dataset('iris')
iris
```

```python
iris['sepal_length'] >= 7
```

```python
# sepal_length의 길이가 7cm 이상인 것만 조회
iris.loc[iris['sepal_length'] >=7, :]  # :의 뜻 -> 모든 데이터를 가져오기  # True값만 출력됨
```

```python
# species가 virginica인 것만 조회
iris.loc[iris['species'] == 'virginica', :]
# iris.loc['컬럼명'] == 'virginica', :]

# 문자열은 'A' == 'B' 로! (=를 두 개 쓴다)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# embark_town 컬럼에서 Southampton 거주민의 모든 데이터를 조회하세요

titanic.loc[titanic['embark_town'] == 'Southampton', :]
```

```python
# age 평균 구하고, 평균 이상인 데이터만 조회하기
titanic['age'].mean()
titanic.loc[titanic['age'] >= 29.69911764705882, :]
```

```python
titanic.loc[titanic['age'] >= titanic['age'].mean(), :]
```

```python
# 이런 방법도 있다(제일 많이 쓰는 방법)
var = 'fare'
titanic.loc[titanic[var] >= titanic[var].mean(), :].head()  # head: 데이터 중 처음 일부분만 반환, (n)에서 n 지정 안하면 기본값 5
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# total_bill의 중간값 이하인 것만 조회
tips.loc[tips['total_bill'] <= tips['total_bill'].median(), :]
```

```python
# day가 Sun인 데이터만 조회하기
tips.loc[tips['day'] == 'Sun',:]
```

```python
# 다중 조건 기본식
# tips.loc[(조건1) & (조건2), :]

# AND 조건 : &
# OR 조건 : |
```

```python
# tip의 평균보다 크면서, time이 Dinner인 데이터만 조회

tips.loc[
    (tips['tip'] >= tips['tip'].mean()) &
    (tips['time'] == 'Dinner'),
:]
```

```python
# 다중조건 예시

tips.loc[(tips['tip'] >= tips['tip'].mean()) &
    (tips['time'] == 'Dinner') &
    (tips['size'] == 3) &
    (tips['sex'] == 'Male') &
    (tips['day'] == 'Sat'),
:]
```
