## 데이터 불러오기

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")
titanic.head()
```

```python
titanic["Name"].str.lower()   # str.lower() : 문자열 열의 모든 문자를 소문자로 변환
```

```python
# 쉼표 이전의 값 추출하기

titanic["Name"].str.split(",")  # str.split() : 문자열을 특정 구분자 기준으로 나누는 문자열 메서드
                                # 쉼표를 기준으로 나눔
```

```python
titanic["Surname"] = titanic["Name"].str.split(",").str.get(0)  # str.get(0) : 각 문자열에서 첫 번째 문자 가져오기
titanic["Surname"]
```

```python
titanic["Name"].str.contains("Countess")  # str.contains() : 특정 문자가 포함되어 있는지에 관한 여부를 출력
```

```python
titanic[titanic["Name"].str.contains("Countess")]  # true인 데이터값만 출력
```

```python
# 가장 긴 문자열 출력
titanic["Name"].str.len()   # str.len() : 각 문자열 길이 반환
```

```python
titanic["Name"].str.len().idxmax()   # idxmax() : 데이터 열에서 최대값을 갖는 인덱스를 반환
```

```python
titanic.loc[titanic["Name"].str.len().idxmax(),"Name"]
```

```python
titanic["Sex_short"] = titanic["Sex"].replace({"male" : "M", "female" : "F"})  # replace : 데이터의 값을 일괄적으로 대체, 교체
titanic["Sex_short"]
```
